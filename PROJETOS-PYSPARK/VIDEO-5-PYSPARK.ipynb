```
# IMPORTANDO BIBLIOTECAS 

import pyspark.sql.functions
from pyspark.sql.types import StructType, StructField , StringType , IntegerType , FloatType
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

# CRIANDO SESSÃO DO SPARK 

spark = (SparkSession.builder
        .master("local(*)")
        .appName('treino1')
        .getOrCreate()
        )


# CRIANDO DATAFRAME 

dados = (

    ['1','José','Silva','Anápolis','São Paulo','01-09-1900'],
    ['2','Igor','Alves','Anápolis','São Paulo','11-09-1977'],
    ['3','Leonardo','Lopes','Anápolis','São Paulo','21-12-2000'],
    ['4','Humberto','Ferreira','Pato Branco','Rio Grande do Sul','13-11-1964'],
    ['5','Isaías','Pereira','Pato Branco','Rio Grande do Sul','07-07-2002'],
    ['6','Lucas','Santos','Tauá','Ceará','05-09-1984']
)


colunas = (StructType([
    StructField('cod' , StringType(), True),
    StructField('nome' , StringType(), True),
    StructField('sobrenome' , StringType(), True),
    StructField('municipio' , StringType(), True),
    StructField('estado' , StringType(), True),
    StructField('data_nasc' , StringType(), True)
]))

## DATAFRAME FINAL 

df = spark.createDataFrame(data = dados , schema = colunas)

df.display()
```

```
# 1) CRIE UMA COLUNA (ANO) - PEGUE SOMENTE O (ANO) DA COLUNA (DATA_NASC) ==> EXISTE BASICAMENTE 3 FORMAS ==> (SPLIT) / (SUBSTRING) / (TO_DATE)

## ==> COM (SPLIT)

#df.withColumn('Ano' , split(col('data_nasc') , '-').getItem(2)).display()

## ==> COM (SUBSTRING)

#df.withColumn('Ano' , substring('data_nasc' , 7 , 4)).display() 


## ==> COM (to_date()) ==> PRIMEIRO TRANSFORMA PARA (DATE) DEPOIS FALA SOMENTE O QUE QUER USANDO (DATE_FORMAT) => ano = 'yyyy'

df = df.withColumn('Ano' , to_date(col('data_nasc'), 'dd-MM-yyyy'))

df = df.withColumn('Ano' , date_format(col('Ano'), 'yyyy'))

df.display()

```

```
# 2) CRIE UMA COLUNA (MES) - PEGUE SOMENTE O (MES) DA COLUNA (DATA_NASC) ==> EXISTE BASICAMENTE 3 FORMAS ==> (SPLIT) / (SUBSTRING) / (TO_DATE)

## ==> COM (SPLIT)

#df.withColumn('Mes' , split(col('data_nasc') , '-').getItem(1)).display()

## ==> COM (SUBSTRING)

#df.withColumn('Mes' , substring('data_nasc' , 4 , 2)).display() 


## ==> COM (to_date()) ==> PRIMEIRO TRANSFORMA PARA (DATE) DEPOIS FALA SOMENTE O QUE QUER USANDO (DATE_FORMAT) => mes = 'MM'

df = df.withColumn('Mes' , to_date(col('data_nasc'), 'dd-MM-yyyy'))

df = df.withColumn('Mes' , date_format(col('Mes'), 'MM'))

df.display()

```

```
# 3) CRIE UMA COLUNA (DIA) - PEGUE SOMENTE O (DIA) DA COLUNA (DATA_NASC) ==> EXISTE BASICAMENTE 3 FORMAS ==> (SPLIT) / (SUBSTRING) / (TO_DATE)

## ==> COM (SPLIT)

#df.withColumn('Dia' , split(col('data_nasc') , '-').getItem(0)).display()

## ==> COM (SUBSTRING)

#df.withColumn('Dia' , substring('data_nasc' , 1 , 2)).display() 


## ==> COM (to_date()) ==> PRIMEIRO TRANSFORMA PARA (DATE) DEPOIS FALA SOMENTE O QUE QUER USANDO (DATE_FORMAT) => dia = 'dd'

df = df.withColumn('dia' , to_date(col('data_nasc'), 'dd-MM-yyyy'))

df = df.withColumn('dia' , date_format(col('dia'), 'dd'))

df.display()

```

```
# 4) NA COLUNA (data_nasc) TROQUE O SIMBOLO (-) por (/) ==> utilizando (REGEXP_REPLACE)


df.withColumn('data_nasc', regexp_replace('data_nasc' , '\-' , '/')).display()
```

# Comparando Intervalos de (Datas) ==> (datediff) / round(months_between)

```
## CRIANDO OUTRO DATAFRAME ==> PARA TRABALHAR COM OUTROS VALORES 

# CRIANDO DATAFRAME

data1 = (["1","2019-04-15"],
       ["2","2020-05-20",],
       ["3","2021-06-25"],
       ["4","2022-06-28"])


schema1 = (StructType([
    StructField('id', StringType(), True),
    StructField('date', StringType(), True)
]))


# DATAFRAME FINAL 

df1 = spark.createDataFrame(data = data1 , schema = schema1)

df1.display()
```

```
# 1) CRIE UMA COLUNA CONTENDO A (DATA ATUAL) E CRIE OUTRA COLUNA CONTENDO A ( (QNT DE DIAS) DE DIFERENCA DA DATA ATUAL ) com a COLUNA (DATE)

df1 = df1.withColumn('Data_atual' , current_date())\
        .withColumn('Dias_corridos', datediff( current_date() , col('date')))

df1.display()

## OBS ==> (datediff) ==> função que calcula os (DIAS)
```

```
# 2) CRIE UMA COLUNA (MESES_CORRIDOS) E CALCULE QUANTOS MESES SE PASSARAM DA COLUNA (DATE) ATE A DATA ATUAL 

df1.withColumn('Meses_corridos' , round(months_between(current_date(), col('date') ),0)).display()
```

```
# 3) CRIE UMA COLUNA (ANOS_CORRIDOS) E CALCULE QUANTOS ANOS SE PASSARAM DA COLUNA (DATE) ATE A DATA ATUAL 

df1.withColumn('Anos_corridos' , round(months_between(current_date() , col('date'))/lit(12), 0)).display()
```

```
# 4) CRIE 3 COLUNAS ==> (DIAS_CORRIDOS) / (MESES_CORRIDOS) / (ANOS_CORRIDOS) ==> 
#UTILIZANDO A COLUNA (DATE) COMO REFERENCIA COM A (DATA ATUAL) 

# (current_date())  ===> data atual
# (datediff)   ===> calcula os dias 
# (round + months , 0)   ===> qnt de meses sem (arredondar)
# (round + months , lit(12),0)   ==> qnt de anos sem (arredondar)
df1 = df1.withColumn('Dias_corridos' , datediff( current_date() , col('date') ))\
        .withColumn('Meses_corridos' , round(months_between( current_date(), col('date')),0))\
        .withColumn('Anos_corridos' , round(months_between(current_date(),col('date'))/lit(12), 0)).display()
```
