```
# IMPORTANDO BIBLIOTECAS 

import pyspark.sql.functions
from pyspark.sql.types import StructType, StructField , StringType , IntegerType , FloatType
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

# CRIANDO SESSÃO DO SPARK 

spark = (SparkSession.builder
        .master("local(*)")
        .appName('treino1')
        .getOrCreate()
        )

```

```
# CRIANDO DATAFRAME 

dados = (

    ['1','José','Silva','Anápolis','São Paulo','01-09-1900'],
    ['2','Igor','Alves','Anápolis','São Paulo','11-09-1977'],
    ['3','Leonardo','Lopes','Anápolis','São Paulo','21-12-2000'],
    ['4','Humberto','Ferreira','Pato Branco','Rio Grande do Sul','13-11-1964'],
    ['5','Isaías','Pereira','Pato Branco','Rio Grande do Sul','07-07-2002'],
    ['6','Lucas','Santos','Tauá','Ceará','05-09-1984']
)


colunas = (StructType([
    StructField('cod' , StringType(), True),
    StructField('nome' , StringType(), True),
    StructField('sobrenome' , StringType(), True),
    StructField('municipio' , StringType(), True),
    StructField('estado' , StringType(), True),
    StructField('data_nasc' , StringType(), True)
]))

## DATAFRAME FINAL 

df = spark.createDataFrame(data = dados , schema = colunas)

df.display()
```

```
# 1) ENCONTRE O ESTADO QUE TEM A LETRA (Ã) E CRIE UMA COLUNA COM A (REGIÃO) DE (SUDESTE)

df = df.withColumn('Região' , when(col('estado').like('%ã%') , lit('Sudeste') ).otherwise(''))

df.display()
```

```
# 2) ENCONTRE O (ESTADO) QUE COMEÇA COM (Rio) E COLOQUE SUA (REGIÁO) COMO = (SUL)


df = df.withColumn('Região' , when(col('estado').like('Rio%') , lit('Sul') ).otherwise(col('Região')))

df.display()
```

```
# 3) ENCONTRE O (ESTADO) QUE TERMINE COM (rá) E COLOQUE SUA (REGIÃO) COMO = (NORDESTE)

df = df.withColumn('Região' , when(col('estado').endswith("rá") , lit('Nordeste') ).otherwise(col('Região')))

df.display()
```

```
# 4) CRIE UMA COLUNA CHAMADA (LOCAL) => COM AS COLUNAS (MUNICIPIO) + (ESTADO) + (REGIAO)

df = df.withColumn('Local' , concat_ws(' - ' , 'municipio','estado','Região' ))

df.display()
```

```
# 5) CRIE UMA COLUNA (ANO) - PEGUE SOMENTE O (ANO) DA COLUNA (DATA_NASC) ==> EXISTE BASICAMENTE 3 FORMAS ==> (SPLIT) / (SUBSTRING) / (TO_DATE)

## ==> COM (SPLIT)

#df.withColumn('Ano' , split(col('data_nasc') , '-').getItem(2)).display()

## ==> COM (SUBSTRING)

#df.withColumn('Ano' , substring('data_nasc' , 7 , 4)).display() 


## ==> COM (to_date()) ==> PRIMEIRO TRANSFORMA PARA (DATE) DEPOIS FALA SOMENTE O QUE QUER USANDO (DATE_FORMAT) => ano = 'yyyy'

df = df.withColumn('Ano' , to_date(col('data_nasc'), 'dd-MM-yyyy'))

df = df.withColumn('Ano' , date_format(col('Ano'), 'yyyy')).display()

```
