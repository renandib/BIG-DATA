```
# FUNÇÕES UTILIZADAS ==> (ISIN) == (FILTER ou where) ==> a diferença é que consegue filtrar mais valores 
```

```
from pyspark.sql import SparkSession
import pyspark.sql.functions 
from pyspark.sql.functions import *
from pyspark.sql.types import StructField , StructType , StringType , IntegerType

# SESSÃO DO SPARK 

spark = (SparkSession.builder
        .appName('pyspark-14')
        .getOrCreate())

dados = (

    ['1','Celta','Chevrolet',2004,18000,"Não"],
    ['2','Gol','volkswagen',2020,46320,"Sim"],
    ['3','Corsa Classic', 'Chevrolet',2008,15000,"Sim"],
    ['4','Palio','Fiat',1998,8900,"Não"],
    ['5','ASX','Mitsubishi',2016,65000,"Sim"],
    ['6','Corolla','Toyota',2018,89990,"Sim"],
    ['7','Civic','Honda',2019,89990,"Sim"],
    ['8','Eco Sport','Ford',2017,69299,"Sim"],
    ['9','Onix','Chevrolet',2020,58990,"Sim"]

)

#SCHEMA

colunas = (StructType([
    StructField('cod' , StringType(), True),
    StructField('Carro' , StringType(), True),
    StructField('Marca' , StringType(), True),
    StructField('Ano' , IntegerType(), True),
    StructField('Preco' , IntegerType(), True),
    StructField('Flex' , StringType(), True)
]))


```

```
# CRIANDO DATA FRAME FINAL 

df = spark.createDataFrame(data = dados , schema = colunas)

df.display()
```

```
df.printSchema()
```

```
## VAMOS UTILIZAR A FUNÇÃO (isin) ==> BASICAMENTE BUSCA VALORES EM CADA COLUNA. (DIGAMOS QUE CADA COLUNA É UMA (LISTA))

# É UTILIZADA , PARA DEIXAR O CODIGO (MENOR) ==> PRATICAMENTE A MESMA FUNCAO DO (FILTER E DO WHERE)
```

```
# 1) FAÇA UMA LISTA COM ESSES VALORES DA COLUNA (MARCAS) => ('Chevrolet', 'Fiat', 'Honda')
# Encontre os dados sobre esses valores da lista
lista = ['Chevrolet','Fiat','Honda'] 

df.filter(col('marca').isin(lista)).display()


```

```
# 2) ENCONTRE CARROS QUE TENHA O (ANO = 2019) 

df.filter(col('Ano') == 2019).display()

#df.filter(col('Ano').isin(2019)).display() (mesma coisa)
```

```
# 3) ENCONTRE OS DADOS DOS CARROS (Civic, Palio)

df.filter(col('Carro').isin('Civic','Palio')).display()
# NÃO FUNCIONA ABAIXO => (O WHERE SÓ PEGA FILTROS DIFERENTES)
#df.filter(col('Carro') == 'Civic').filter(col('Carro') == 'Palio').display()
```

```
# 4) ENCONTRE CARROS DA (MARCA = CHEVROLET) QUE (NÃO) SEJA (FLEX)

df.filter(col('Marca') == 'Chevrolet').filter(col('Flex') == 'Não').display()

#abaixo tambem da 
#df.filter(col('Marca').isin('Chevrolet')).filter(col('Flex').isin('Não')).display()
```

```
# 5) ENCONTRE CARROS DA (MARCA = Fiat , Honda) QUE (NÃO) SEJAM FLEX

df.filter(col('Marca').isin('Fiat','Honda')).filter(col('Flex') == 'Não').display()

"""
ABAIXO NÃO FUNCIONA
df.filter(col('Marca') == 'Fiat')\
    .filter(col('Marca') == 'Honda')\
    .filter(col('Flex') == 'Não').display()
"""
```

```
# 6) MOSTRE TODOS OS CARROS , 
#QUE (NÃO) SÃO DA (MARCA = Fiat , Honda ,Chevrolet)
#utilizando (~ isin) = (~) = 'NÃO'

df.filter(~ col('Marca').isin('Fiat','Chevrolet','Honda')).display()
```
