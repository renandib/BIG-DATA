```
# IMPORTANDO BIBLIOTECAS 

import pyspark.sql.functions
from pyspark.sql.types import StructType, StructField , StringType , IntegerType , FloatType
from pyspark.sql import SparkSession
from pyspark.sql.functions import *


# CRIANDO SESSÃO DO SPARK 

spark = (SparkSession.builder
        .master("local(*)")
        .appName('VIDEO3')
        .getOrCreate()
        )

# CRIANDO DATAFRAME 

dados = (

    ['James','Vendas','NY',90000,34,10000],
    ['Michael','Vendas','NY',86000,56,20000],
    ['Robert','Vendas','CA',81000,30,23000],
    ['Maria','Financas','CA',90000,24,23000],
    ['James','Vendas','CA',99000,40,24000],
    ['Scott','Financas','NY',83000,36,19000],
    ['Jonatas','Financas','NY',79000,53,15000],
    ['Jefferson','Marketing','NY',79000,53,15000],
    ['Caio','Marketing','CA',80000,25,18000],
    ['Lucio','Vendas','NY',91000,50,21000]
)


colunas = (StructType([
    StructField('Nome' , StringType(), True),
    StructField('Departamento' , StringType(), True),
    StructField('Estado' , StringType(), True),
    StructField('Salario' , IntegerType(), True),
    StructField('Idade' , IntegerType(), True),
    StructField('Bonus' , IntegerType(), True)
    
]))

## DATAFRAME FINAL 

df = spark.createDataFrame(data = dados , schema = colunas)

df.display()
```

```
# 1) Organize em ordem crescente a coluna (Departamento)   ==> sort = orderBy

#df.orderBy('Departamento').display() # => tbm funciona assim ==> df.sort(....)

df.orderBy(col('Departamento').asc()).display()
```

```
# 2) Organize a coluna (Idade) por ordem  (Decrescente)

df.orderBy(col('Idade').desc()).display()
```

```
## AGORA VAMOS TRABALHAR COM (AGG) ==> AGREGAÇÕES ==> (MAX, MIN , SUM , AVG)

# 3) Agrupe a coluna (Departamento) e faça a (Soma Total) dos salários de cada Departamento. 

#df.groupBy('Departamento').sum('Salario').display()

df.groupBy('Departamento').agg(sum('Salario').alias('Total-salarios')).display()
```

```
# 4) Informe a (quantidade) de cada (Departamento) 

df.groupBy('Departamento').agg(count('Departamento').alias('N.Depart')).display()
```

```
# 5 ) INFORME A (SOMA DE SALARIOS) E A (SOMA DE BONUS) POR (DEPARTAMENTO) E TAMBEM POR (ESTADO)


df.groupBy('Departamento','Estado').agg(sum('Salario').alias('S.salario'), sum('Bonus').alias('S.bonus')).display()
```

```
# 6) INFORME da coluna (Salario) a (SOMA TOTAL) ; (MÉDIA) ; (MIN) E (MAX) da coluna (Departamento)

df.groupBy('Departamento').agg(sum('Salario').alias('S.salario'),\
                               avg('Salario').alias('Media.salario'),\
                               min('Salario').alias('Menor.salario'),\
                               max('Salario').alias('Maior.salario')\
                              ).display()
```

```
# 7) Faça a mesma coisa acima (6) ==> Porem, AONDE O (BONUS) SEJA MAIOR QUE (21000)

df.filter(col('Bonus') >= 21000).groupBy('Departamento').agg(sum('Salario').alias('Soma.salario'),\
                              avg('Salario').alias('Media.salario'),\
                              min('Salario').alias('Menor.salario'),\
                              max('Salario').alias('Maior.salario')).display()
                              
```
