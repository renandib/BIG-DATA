```
# IMPORTANDO BIBLIOTECAS 

import pyspark.sql.functions
from pyspark.sql.types import StructType, StructField , StringType , IntegerType , FloatType
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

# CRIANDO SESSÃO DO SPARK 

spark = (SparkSession.builder
        .master("local(*)")
        .appName('video6')
        .getOrCreate()
        )

# CRIANDO DATAFRAME 

dados = (

    ['1','José','Silva','Anápolis','São Paulo','01-09-1900'],
    ['2','Igor','Alves','Anápolis','São Paulo','11-09-1977'],
    ['3','Leonardo','Lopes','Anápolis','São Paulo','21-12-2000'],
    ['4','Humberto','Ferreira','Pato Branco','Rio Grande do Sul','13-11-1964'],
    ['5','Isaías','Pereira','Pato Branco','Rio Grande do Sul','07-07-2002'],
    ['6','Lucas','Santos','Tauá','Ceará','05-09-1984']
)


colunas = (StructType([
    StructField('cod' , StringType(), True),
    StructField('nome' , StringType(), True),
    StructField('sobrenome' , StringType(), True),
    StructField('municipio' , StringType(), True),
    StructField('estado' , StringType(), True),
    StructField('data_nasc' , StringType(), True)
]))

## DATAFRAME FINAL 

df = spark.createDataFrame(data = dados , schema = colunas)

df.display()
```

```
# 1) CRIE UMA COLUNA (IDADE) A PARTIR DA (data_nasc) 

df = df.withColumn('Idade' , date_format(current_timestamp(), 'yyyy')\
            .cast('int') - substring(col('data_nasc') , 7, 4)\
             .cast('int'))
df.display()
```

```
# OBS 1 ) lpad( col('coluna') , tamanho ocupado ,  'o que sera ADD') )

# OBS 2 ) LPAD = ADD A (ESQUERDA)  ;  RPAD = ADD A (DIREITA)

# 2) NA COLUNA (COD) add (2x) a (ESQUERDA) do (cod)   

df = df.withColumn('cod' , lpad(col('cod') , 3 , 'x'))

# 3) NA COLUNA (Idade) add ('anos') a (DIREITA) da (idade)

df = df.withColumn('Idade' , rpad(col('Idade') , 8 , ' anos'))

df.display()
```

# Unindo colunas (sem o mesmo schema)  ==> (concat_ws) / (overlay)

```
# 4 ) CRIE UMA COLUNA (NOME_COMPLETO) (UNINDO ) AS COLUNAS (nome) + (sobrenome) ===> USANDO (concat)

df = df.withColumn('Nome_completo' , concat_ws(' - ' , 'nome' , 'sobrenome')) # Consegue (ADD) (ELEMENTOS)

df = df.withColumn('Nome_compl2' , concat('nome' , 'sobrenome')) # não consegue (add) (ELEMENTOS)

df.display()
```

```
# 5) CRIE UMA COLUNA (COD_NOME) (UNINDO) AS COLUNAS (cod + nome) ===> USANDO (overlay) ==> MESMA COISA QUE O (CONCAT) => (POREM, PRECISA COLOCAR O TAMANHO OCUPADO)

df = df.withColumn('cod_nome' , overlay('cod' , 'nome' , 11)).display()
```
